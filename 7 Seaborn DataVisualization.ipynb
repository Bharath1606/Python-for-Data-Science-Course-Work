```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# plot the distribution of the data
sns.distplot(iris['petal_length'], bins = 20, kde=False)
plt.show()
```

```python
sns.kdeplot(iris['petal_length'])
plt.show()
```

```python
sns.scatterplot(x=iris['petal_length'], y= iris['petal_width'])
plt.show()
```

```python
sns.scatterplot(x = 'petal_length', y= 'petal_width',  data = iris, hue = 'species')
plt.show()
```

```python
sns.jointplot(x= 'sepal_length', y= 'sepal_width',data = iris, kind = 'reg')
plt.show()
```

# Pair Plot

```python
sns.pairplot(data = iris, hue = 'species')
```

# Categorical Plots

```python
sns.countplot(x = 'species', data = iris)
plt.show()
```

```python
plt.figure(figsize = (10,10))
sns.boxplot(x = 'species', y = 'petal_length', data = iris)
```

```python
sns.violinplot(x= 'species', y= 'petal_length', data = iris)
plt.show()
```

### Finding correlation

```python
iris.corr()
```

```python
sns.scatterplot(x = 'sepal_length', y = 'sepal_length', data = iris)
```

```python
sns.heatmap(iris.corr(), cmap = 'Blues',annot =True)
plt.show()
           
```

# Visualise Tips Dataset

```python
tips = sns.load_dataset('tips')
```

```python
tips.shape
```

```python
tips.head()
```

```python
sns.countplot(x= tips['sex'])
plt.show()
```

```python
tips['day'].unique()
```

```python
# out of all the days which day gets the highest amount
# Ans = Sunday

sns.barplot(x= 'day', y = 'total_bill', data = tips, estimator=np.mean, hue = 'sex')
plt.show()
```

```python
# what do people prefer - lunch or dinner

sns.barplot(x ='time', y = 'total_bill', data=tips, hue = "day")
plt.show()
```

```python
# Do smoker people spend more money?

sns.barplot(x = 'smoker', y = 'tip', data = tips)
plt.show()
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips)
plt.show()
```
