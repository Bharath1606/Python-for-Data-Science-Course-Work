```python
import matplotlib.pyplot as plt
import numpy as np
```

# Line Plots

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x+3
```

```python
plt.plot(x, y1)
plt.plot(x,y2)

plt.show()
```

```python
themes = plt.style.available
print(themes)
```

```python
plt.style.use('seaborn')
```

```python
plt.plot(x, y1, color = 'red', marker= 'o', label ="iphone")
     
plt.plot(x,y2, color= 'green', marker='*', markersize = 12, linestyle = "dashed", label ="Anroid")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Pices of smartphone over time")
plt.legend()
plt.show()
```
