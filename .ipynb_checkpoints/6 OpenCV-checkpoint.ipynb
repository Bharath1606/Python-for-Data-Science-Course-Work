# Image Rotation

```python
img_rgb.shape
```

```python
rows, cols = img_rgb.shape[:-1]
```

```python
M = cv2.getRotationMatrix2D((rows/2, cols/2), 10, 0.9)
                       
```

```python
i = cv2.warpAffine(img_rgb, M, (cols, rows))
```

```python
plt.imshow(i)
          
```

# Get Video From Camera

```python
import cv2
```

```python
%matplotlib
```

```python
cam = cv2.VideoCapture(0)
```

```python
while True:
    ret,frame = cam.read()
    
    if ret == False:
        continue
        
    cv2.imshow("My frame", frame)
    
    # waiting for user to press a key
    # if user presses 'q' then terminate
    key_pressed = cv2.waitKey(1) & 0xFF
    if key_pressed == ord('q'):
        break
        
        
cam.release()
cv2.destroyAllWindows()
```

# Face Detection

```python
import cv2
import matplotlib.pyplot as plt
```

```python
import os
os.chdir('/Users/bharathtallapureddy/Downloads/ DSMC/PDMC-2021/Computer vision with Open CV')
```

```python
img = plt.imread("./group-photo.jpg")
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
# Loading the model
detector = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
```

```python
all_faces = detector.detectMultiScale(img, 1.2, 5)
```

```python
all_faces.shape
```

```python
x,y,w,h = all_faces[0]
```

```python
img = cv2.rectangle(img, (x,y),(x+w, y+h), (255,0,0,), 5)
```

```python
plt.imshow(img)
```

```python
# Plotting rectangle for all faces
for face in all_faces:
    x,y,w,h = face
    img = cv2.rectangle(img, (x,y),(x+w, y+h), (255,0,0,), 5)
```

```python
plt.imshow(img)
```

# Face Detection in Real Time

```python
import cv2
```

```python
%matplotlib
```

```python
detector = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
```

```python
cam = cv2.VideoCapture(0)
while True:
    ret,frame = cam.read()
    
    if ret == False:
        continue
        
    all_faces = detector.detectMultiScale(frame, 1.5, 5)
    
    for face in all_faces:
        x,y,w,h = face
        frame = cv2.rectangle(frame, (x,y),(x+w, y+h), (0,0,255), 5)
    
    # this will show the frame
    cv2.imshow("Face Detection", frame)
    
    # finding which key yo've pressed
    key_pressed = cv2.waitKey(1) & 0xFF
        
    if key_pressed == ord('q'):
            break
        
        
cam.release()
cv2.destroyAllWindows()
    
```
