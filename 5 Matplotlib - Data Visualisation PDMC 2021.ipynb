# Movie Data Visualization
#Given mlvie Dataset, your task is to make a visualization that-

-Plot the lenght of movie title names on x-Axis,

-Frequency counts on Y-axis i.e Number of movies having x characters in their title

```python
movie = pd.read_csv('./movie_meta.csv')
```

```python
movie.head()
```

```python
movie.info()
```

```python
movie.columns
```

```python
def preprocess(ele):
    return len(ele.strip())
```

```python
movie['movie_title_len'] = movie['movie_title'].apply(preprocess)
```

```python
movie.columns
```

```python
plt.hist (movie['movie_title_len'],bins = 50)
plt.xticks(range(0,100,5))
plt.xlabel("Movie title len")
plt.ylabel("Number of movies (Freq)")
plt.show()
```

### Scatter way

```python
movie['movie_title_len']
```

```python
freq_title = {}

for l in movie ['movie_title_len']:
    if freq_title.get(l) is None:
        freq_title[l] = 1
    else:
        freq_title[l] +=1
```

```python
x = freq_title.keys()
y = freq_title.values()
```

```python
plt.scatter(x,y)
plt.xticks(range(0,60,5))
plt.xlabel("Movie title len")
plt.ylabel("Number of movies (Freq)")
plt.show()
```
